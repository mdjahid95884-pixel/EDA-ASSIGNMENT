Question 1 : What is the difference between AI, ML, DL, and Data Science? Provide a brief explanation of each.

- Artificial Intelligence (AI) is the broad idea of building machines that can perform tasks that normally require human intelligence—like reasoning, problem-solving, and decision-making.
Machine Learning (ML) is a subset of AI where the machine learns patterns from data instead of being manually programmed for every rule.
Deep Learning (DL) is a specialized part of ML that uses neural networks with many layers, allowing the system to learn very complex patterns (like images, audio, or natural language).
Data Science, on the other hand, is a larger ecosystem focused on extracting insights from data using statistics, ML, visualization, and domain knowledge
AI = Goal (making machines intelligent)
ML = Method (learning from data)
DL = Advanced ML (neural networks)
Data Science = Understanding & solving problems using data

Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent them?

- Overfitting happens when a model memorizes the training data so well that it fails to generalize to unseen data. You will notice the training accuracy is high while test accuracy is poor.Underfitting is the opposite—the model is too simple and performs poorly on both training and testing data

You can detect them through
A clear gap between training and validation loss
Very high variance (overfitting) or very high bias (underfitting)

To prevent them
Use cross-validation to check model performance on multiple splits
Use regularization (L1/L2) to reduce model complexity
Use dropout (for neural networks)
Collect more data or perform data augmentation

Question 3:How would you handle missing values in a dataset? Explain at least three methods with examples.

- Real-world datasets almost always contain missing values. A few practical ways to handle them are

Deletion
If a row has too many missing values, removing it is often the easiest method.
Example: Dropping rows where 70% of columns are empty

Mean/Median Imputation
Replace missing numerical values with the mean or median of that column.
Example: Missing age values in a dataset can be filled with median age.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it(theoretical + practical).

- An imbalanced dataset is one where one class appears far more frequently than the other.
Example: Fraud detection datasets often have 98% “non-fraud” and only 2% fraud cases

Techniques to handle imbalance

SMOTE (Synthetic Minority Oversampling Technique)
SMOTE creates synthetic data points for the minority class, helping the model learn patterns more effectively.
Class weights in models
Many algorithms (like Logistic Regression or Random Forest) allow you to increase penalty for misclassifying minority classes.
This improves recall for the minority group

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.


- Feature scaling ensures that all features contribute equally to the model.
Algorithms like KNN, SVM, and gradient descent perform poorly when features are on different scales.

Min-Max Scaling (Normalization)
Brings values into a fixed range (0 to 1).
Useful when you need bounded values.

Standardization (Z-score scaling)
Converts values to mean = 0 and standard deviation = 1.
Works better when data contains outliers.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

- Label Encoding assigns numbers to categories (e.g., Low = 1, Medium = 2 High = 3).

Best for ordinal data where the order matters one-Hot Encoding creates separate binary columns for each category.
Useful for nominal data like colors or cities, where categories don’t have any order.

```python
""" Question 7: Google Play Store Dataset
    a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.) """

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("googleplaystore.csv")
df.head()


```

```python
category_ratings = df.groupby("Category")["Rating"].mean().sort_values(ascending=False)
print(category_ratings)
```

Based on the average ratings

Top-rated categories were generally Education, Books, and Health & Fitness.
These categories usually offer useful, low-bug apps, which reflects in higher user satisfaction.

Lowest-rated categories tended to be Dating, Finance, and Lifestyle.
Users often complain about too many ads, bugs, or paywalls, leading to poor ratings.

```python
#Question 8: Titanic Dataset

# a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?

df = pd.read_csv("titanic.csv")
df.head()

```

```python
#Question 8: Titanic Dataset
# b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?

df["Child"] = df["Age"] < 18
child_survival = df.groupby("Child")["Survived"].mean()
print(child_survival)
```

Children did, in fact, have a better chance of survival.
